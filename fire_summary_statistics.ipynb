```python
import pandas as pd
```

```python
df = pd.read_csv ('FirePy/Fire.csv')

df
```

```python
avg = df.groupby(['STRUCTURE TYPE']).mean()["PERCENTAGE LOSS"]
median = df.groupby(['STRUCTURE TYPE']).median()["PERCENTAGE LOSS"]
var = df.groupby(['STRUCTURE TYPE']).var()["PERCENTAGE LOSS"]
std = df.groupby(['STRUCTURE TYPE']).std()["PERCENTAGE LOSS"]
sem = df.groupby(['STRUCTURE TYPE']).sem()["PERCENTAGE LOSS"]


percent_statistics = pd.DataFrame({ 
                             "Tumor Volume Mean": avg,
                             "Tumor Volume Median": median,
                             "Tumor Volume Var": var,
                             "Tumor Volume Std": std,
                             "Tumor Volume Sem": sem
    
                         
                        })
percent_statistics
```
